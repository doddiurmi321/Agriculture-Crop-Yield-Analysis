```python
import pandas as pd
```

```python
df = pd.read_csv('/content/crop_yield.csv')
```

```python
import numpy as np
```

```python
yield_array = np.array(df['Yield'])
print(yield_array)
```

```python
np.mean(yield_array)
```

```python
np.max(yield_array)
```

```python
np.min(yield_array)
```

```python
rainfall_array = np.array(df['Annual_Rainfall'])
print(rainfall_array)
```

```python
production_array = np.array(df['Production'])
print(production_array)
```

```python
area_array = np.array(df['Area'])
print(area_array)
```

```python
print("Yield Standard Deviation:", round(np.std(yield_array), 2))
```

```python
efficiency = production_array / area_array
print("Production Efficiency (first 5):", efficiency[:5])
```

```python
high_yield = yield_array[yield_array > np.mean(yield_array)]
print("High Yield Values:", high_yield[:10])
```

```python
index = np.argmax(yield_array)
print("Best Yield Index:", index)
print("Best Yield Value:", yield_array[index])
```

```python
print("25% Yield:", np.percentile(yield_array, 25))
print("50% Yield (Median):", np.percentile(yield_array, 50))
print("75% Yield:", np.percentile(yield_array, 75))
```
